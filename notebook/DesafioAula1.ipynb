```python
# Importar bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Carregar .csv
data = pd.read_csv('https://github.com/aeroadriano/AluraBootcamp-DataScience/blob/main/data/A164817189_28_143_208_processamento.csv?raw=True', 
                    encoding='ISO-8859-1',
                    sep=';', decimal=',', thousands='.',
                    skiprows=3, skipfooter=12,
                    engine='python')
```

```python
# Pré-visualização dos dados
data.head()
```

```python
ax = data.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")
plt.show()
```

# Desafio: escolher um título mais descritivo, que passe a mensagem adequada.
# Desafio: faça a mesma análise pro mês mais recente.

```python
recent = data.columns[-2] # Seleciona o mês mais recente
data["Unidade da Federação"] = data["Unidade da Federação"].str.slice(3,) # Retira os números dos nomes das UFs

# Dados do plot
ax = data.sort_values(by=recent,ascending=[0]).plot(x='Unidade da Federação', y=recent, kind='bar', figsize=(15,8), color='darkorange')

# Formata eixo, título e labels
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.title("Total de gastos com procedimentos hospitalares no SUS", fontsize=20, fontweight='bold')
ax.set_xlabel("Unidades da Federação (UF)", fontsize=14)
ax.set_ylabel("Gastos [R$]", fontsize=14)
ax.grid(axis = 'y', color='lightgrey')

# Renderiza
plt.show()
```
